```python
#importing basic Libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
#creating synthetic dataset
from sklearn.datasets import make_classification
x,y=make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=1,n_redundant=0)
```

```python
pd.DataFrame(x)[0]
```

```python
y
```

```python
sns.scatterplot(x=pd.DataFrame(x)[0],y=pd.DataFrame(x)[1],hue=y)
```

```python
#splitting dataset into train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

```python
# moedel training
from sklearn.svm import SVC
svc=SVC(kernel='linear')
svc.fit(x_train,y_train)
```

```python
y_pred=svc.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

Note:- acuuracy is 100% because datapoints are linearly sepearable compltely
 

 we use kernels in case of non linear datapoints

```python
x,y=make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=1,n_redundant=0)
```

```python
sns.scatterplot(x=pd.DataFrame(x)[0],y=pd.DataFrame(x)[1],hue=y)
```

Note here daatpoints are not linearly seperable


```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42)
svc=SVC(kernel='linear')
svc.fit(x_train,y_train)
```

```python
y_pred=svc.predict(x_test)
```

```python
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

Note we can see here using linear kernel accuracy is 81% only

lets use differnt kerel and check accuracy

```python
#polynomial kernel
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
svc=SVC(kernel='poly')
svc.fit(x_train,y_train)
```

```python
y_pred=svc.predict(x_test)
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
# RBf kernel
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
svc=SVC(kernel='rbf')
svc.fit(x_train,y_train)
y_pred=svc.predict(x_test)
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
# sigmoid kernel
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
svc=SVC(kernel='sigmoid')
svc.fit(x_train,y_train)
y_pred=svc.predict(x_test)
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

We can see rbf kernel has most accuracy i:e 89%

```python
#sigmoid kernel
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
svc=SVC(kernel='sigmoid')
svc.fit(x_train,y_train)
y_pred=svc.predict(x_test)
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

## HyperParameter tunning 

```python
from sklearn.model_selection import GridSearchCV
param = {'C':[0.1,1,10,100,1000],'gamma':[1,0.1,0.01,0.001,0.0001],'kernel':['rbf']}
```

```python
grid=GridSearchCV(estimator=svc,param_grid=param,cv=5,verbose=3,refit=True)
```

The verbosity of randomized search in scikit-learn can be controlled using the verbose parameter. The higher the value, the more messages will be printed during the fitting process. 

Refit=True means that it trains on the entire training set after the cross validation is done.

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred=grid.predict(x_test)
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

Using RandamisedSearchCv

```python
from sklearn.model_selection import RandomizedSearchCV
param = {'C':[0.1,1,10,100,1000],'gamma':[1,0.1,0.01,0.001,0.0001],'kernel':['rbf']}
```

```python
randomised=RandomizedSearchCV(estimator=svc,param_distributions=param,cv=5,verbose=3,refit=True)
```

```python
randomised.fit(x_train,y_train)
```

```python
randomised.best_params_
```

```python
randomised.best_score_
```

```python
y_pred=randomised.predict(x_test)
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
