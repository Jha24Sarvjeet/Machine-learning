```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#loading dataset
from sklearn.datasets import load_diabetes
diabetes=load_diabetes()
```

```python
print(diabetes['DESCR'])
```

```python
diabetes
```

```python
#depemndent and dependent features
x=pd.DataFrame(diabetes['data'],columns=['age','sex','bmi','bp','s1','s2','s3','s4','s5','s6'])
y=diabetes['target']
```

```python
x.head()
```

```python
y
```

```python
#splitting train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
x_train.corr()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(x_train.corr(),annot=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor()
regressor.fit(x_train,y_train)
```

```python
y_pred=regressor.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,root_mean_squared_error,r2_score
print(r2_score(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(root_mean_squared_error(y_test,y_pred))
```

```python
from sklearn import tree
plt.figure(figsize=(30,20))
tree.plot_tree(regressor,filled=True)
```

# Hyperparameter tuning

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
param={
    'criterion':['squared_error','friedman_mse','absolute_error'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5,10,15,20,25],
    'max_features':['auto','sqrt','log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid=GridSearchCV(estimator=regressor,param_grid=param,cv=5,scoring='neg_mean_squared_error')
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred=grid.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,root_mean_squared_error,r2_score
print(r2_score(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(root_mean_squared_error(y_test,y_pred))
```

```python
best_model=DecisionTreeRegressor(criterion= 'squared_error',
 max_depth= 2,
 max_features= 'log2',
 splitter= 'best')
```

```python
best_model.fit(x_train,y_train) 
```

```python
plt.figure(figsize=(30,20))
tree.plot_tree(best_model,filled=True)
```
