```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#creating synthetic dataset
from sklearn.datasets import make_classification
x,y=make_classification(n_samples=1000,n_features=3,n_redundant=1,n_classes=2,random_state=999)
```

```python
x
```

```python
sns.displot(x[0],kind='kde')
sns.displot(x[1],kind='kde')
sns.displot(x[2],kind='kde')
```

Note :- faetures are almost following Normal distribution

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
#model training
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5,algorithm='auto')
```

```python
knn.fit(x_train,y_train)
```

```python
y_pred=knn.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
