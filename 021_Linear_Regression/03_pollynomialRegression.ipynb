```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#creating Dataframe
x=6*np.random.randn(100,1)-3
y=0.5 * x**2 + 1.5*x + 2 + np.random.randn(100,1)
## quadratic equation used- y=0.5x^2+1.5x+2+outliers
plt.scatter(x,y,c='g')
plt.xlabel('x')
plt.ylabel('y')
```

```python
sns.displot(x,kind='kde')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
#standardizing
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
print(np.mean(x_train).round(decimals=6))  #checling the  mean after standarization
np.std(x_train).round(decimals=6)  #checling the  satndard deevitaion after standarization
```

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
regression.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score
rSquare=r2_score(y_test,regression.predict(x_test))
rSquare
```

```python
plt.scatter(x_train,y_train,c='r')
plt.plot(x_train,regression.predict(x_train))
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2,include_bias=True)
x_train_poly=poly.fit_transform(x_train)
x_test_poly=poly.transform(x_test)
```

```python
x_train_poly
```

```python
x_test_poly
```

```python
regression.fit(x_train_poly,y_train)
y_pred=regression.predict(x_test_poly)
rScore=r2_score(y_test,y_pred)
rScore
```

```python
plt.scatter(x_train,y_train)
plt.scatter(x_train,regression.predict(x_train_poly),c='r')
```
