Data Visualization With Seaborn

Seaborn is a Python visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn helps in creating complex visualizations with just a few lines of code. In this lesson, we will cover the basics of Seaborn, including creating various types of plots and customizing them.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Basic plotting with seaborn

tips_dataset=sns.load_dataset('tips')
print(tips_dataset)
```

```python
#creating a scatter plot
sns.scatterplot(x='total_bill',y='tip',data=tips_dataset)
plt.title("Scatter plot of Total bills and tips")
```

```python
#line plot
sns.lineplot(x='size',y='total_bill',data=tips_dataset)
plt.title("Line plot of Total bills and size")
```

```python
# bar plot
sns.barplot(x='day',y="total_bill",data=tips_dataset)
```

```python
#box plot
sns.boxplot(x='day',y="total_bill",data=tips_dataset)
```

```python
#violin plot
sns.violinplot(x='day',y="total_bill",data=tips_dataset)
```

```python
#Histogram
sns.histplot(tips_dataset["total_bill"],bins=10)
```

```python
#Histogram
sns.histplot(tips_dataset["total_bill"],bins=10,kde=True)
```

```python
#kde plot
sns.kdeplot(tips_dataset["total_bill"],fill=True)
```

```python
#pair plot
sns.pairplot(tips_dataset)
```

# heatmap

```python
#heatmap
#correlation
corr=tips_dataset[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```

```python
import pandas as pd
sales_df=pd.read_csv('sales_data.csv')
sales_df.head()
```

```python
#plot total sales by product
plt.figure(figsize=(10,6))
sns.barplot(x='Product Category',y='Total Revenue',data=sales_df,estimator=sum)
plt.xlabel('product')
plt.ylabel('Total sales')
plt.title("Total sales by product")
```

```python
#plot total sales by product
plt.figure(figsize=(10,6))
sns.barplot(x='Region',y='Total Revenue',data=sales_df,estimator=sum)
plt.xlabel('Region')
plt.ylabel('Total sales')
plt.title("Total sales by Region")
```
